## 3. ***Data Wrangling***

### Data Wrangling Code

Missing values

'Company' and 'agent' columns have a very high amount of missing values (94,31% and 13,69%, respectively). One possible option could be to drop down that columns. However, the original article (that can be found here) states the following:

In some categorical variables like Agent or Company, “NULL” is presented as one of the categories. This should not be considered a missing value, but rather as “not applicable”. For example, if a booking “Agent” is defined as “NULL” it means that the booking did not came from a travel agent.

Then, I'm going to replace null values in those columns with 0 (as they have float64 data types, since the personal information was removed).

On the other hand, the column 'country' has 0.41% missing values (488 rows affected). In its case, being a categorical variable, I have chosen the mode to replace the null values.

The column 'children' also has some missing values, but they do not even represent 0.01% of the dataset. Therefore, I have chosen to delete the affected rows (4 rows).

````python
# Write your code to make your dataset analysis ready.
# Replace Null values in company and agent columns with 0
df1['company'].fillna(0, inplace=True)
df1['agent'].fillna(0, inplace=True)

# Replaceing Null vaalues in country column with others
df1['country'].fillna('others', inplace=True)

# Remove row aggect by null values in children column
df1.dropna(subset=['children'], inplace=True)
````

````python
# creating a boxplot and historaam for outlier detection
n = 1

sns.set_style('darkgrid')
sns.set(font_scale=1.2)
plt.figure(figsize=(30, 46))

for feature in Numerical_columns:
    plt.subplot(21, 2, n)
    sns.boxplot(df1[feature], palette='autumn').set(
        xlabel='')  # Change None to ''
    plt.title(f'{feature} boxplot')
    n = n + 1

    plt.subplot(21, 2, n)
    plt.hist(df1[feature], color='#f7b267')
    plt.title(f'{feature} histogram')
    n = n + 1
    plt.tight_layout()

plt.show()
````

After analyzing the data provided, it can be observed that there are certain variables that show outliers or unusual values. For instance, in the lead_time column, the data starts to become more dispersed starting at the 400 value. Similarly, the stays_in_weekend_nights and stays_in_week_nights columns show higher outliers for the resort than for the city hotel, although both hotels start to show more dispersed data at 6 and 10 days, respectively.

Regarding the number of adults per reservation, it is clear that the outliers are only for the resort, which could be due to the fact that the resort offers services to large groups of guests. However, without knowing the type of rooms the hotel has, it is difficult to reach a conclusion on this basis. In any case, it is unusual for any reservation to exceed four adults.

In the children and babies columns, there are three data points with very high values, which could be an error when entering the booking data. Similarly, it is extremely unusual for a booking to make more than 5 changes.

The adr variable has an outlier with a particularly high value for the city hotel, which could be an error and should be removed. On the other hand, the days_in_waiting_list column shows sparse data, being unusual for a reservation to have a waiting period of more than one day.

Finally, in the previous_cancellations and previous_bookings_not_canceled columns, it can be observed that these variables reach very high values. While it is true that some of these reservations were made by companies that regularly use these hotels to accommodate their employees, there are also bookings made by individuals where these variables reach high values.

In conclusion, it is important to analyze these variables carefully to ensure that the data entered is accurate and reliable.
